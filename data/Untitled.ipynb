```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('img0_00001.jpg') 

# Convert the image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blur = cv2.GaussianBlur(image_gray, ksize=(9, 9), sigmaX=0)

# Apply thresholding
ret, thresh1 = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)

# Split channels and merge for correct color display in matplotlib
b, g, r = cv2.split(image)
image_rgb = cv2.merge([r, g, b])

# Display the color image
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.imshow(image_rgb)
plt.title('Color Image')
plt.xticks([])
plt.yticks([])

# Display the grayscale image
plt.subplot(132)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.xticks([])
plt.yticks([])

# Display the thresholded image
plt.subplot(133)
plt.imshow(thresh1, cmap='gray')
plt.title('Thresholded Image')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
import cv2

src = cv2.imread("img0_00001.jpg", cv2.IMREAD_COLOR)
src = cv2.resize(src, dsize=(1000,1000 ))

hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

h = cv2.inRange(h, 8, 20)
orange = cv2.bitwise_and(hsv, hsv, mask = h)
orange = cv2.cvtColor(orange, cv2.COLOR_HSV2BGR)

cv2.imshow("orange", orange)
cv2.waitKey()
cv2.destroyAllWindows()
```
